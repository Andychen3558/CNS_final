```python
import json
import gensim
```

```python
wordfile = 'wiki.zh.vec'
m = gensim.models.KeyedVectors.load_word2vec_format(wordfile)
```

```python
import jieba.posseg as pseg
keys = list(m.vocab.keys())
wlist = []
for key in keys:
    words = pseg.cut(key)
    for w, f in words:
        wlist.append([w, f])
```

```python
def filter(wlist):
    contain = ["n"]
    after = []
    for i in wlist:
        if i[1] in contain and i[0] in m.vocab:
            after.append(i[0])
    return set(after)
```

```python
nlist = list(filter(wlist))
```

```python
len(nlist)
```

```python
n = [(m.similarity(i, '學校'), i) for i in nlist]
```

```python
n.sort(key=lambda i: i[0], reverse=True)
```
